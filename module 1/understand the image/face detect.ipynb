# Import The Library

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('male1.jpg')
```

```python
img.shape
```

# Convert The Gray Scale


```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap = 'gray')
```

# Make A Rectangle

```python
haar = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces = haar.detectMultiScale(gray) # tell about the face coordinates
print(faces)
```

```python
cv2.rectangle(img, (41,51), (41+175 , 51+175),(0,255,0),3)
```

```python
plt.imshow(img)
```

# Face Detection

```python
cv2.imshow('detection',img)
```

```python
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
face_crop = img[50:50+261, 60:60+261]
```
