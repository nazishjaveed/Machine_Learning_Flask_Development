```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img_mat = mpimg.imread('img4.jpg') #RGB
import cv2
```

```python
img_cv = cv2.imread('./data/img4.jpg') #BGR
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.imshow(img_mat)
```

```python
r = img_mat[:,:,0] # red channel array
g = img_mat[:,:,1] # green channel array
b = img_mat[:,:,2] # blue channel array
```

```python
r.shape,g.shape, b.shape
```

```python
plt.imshow(b, cmap = 'gray') # bright the different color image through cmap
```

```python
# Rgb image convrt into gray_scale and also write their tittle with original image
```

```python
plt.figure(figsize = (10,6))
plt.subplot(1,4,1)
plt.title('Red Region')
plt.imshow(r, cmap = 'gray')
plt.subplot(1,4,2)
plt.title('Green Region')
plt.imshow(g, cmap = 'gray')
plt.subplot(1,4,3)
plt.title('Blue Region')
plt.imshow(b, cmap = 'gray')
plt.subplot(1,4,4)
plt.title('original image')
plt.imshow(img_mat)
plt.show()
```

```python
gray_mat = cv2.cvtColor(img_mat,cv2.COLOR_RGB2GRAY)
gray_cv = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
```
